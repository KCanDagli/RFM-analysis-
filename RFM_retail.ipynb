# Calculating RFM METRICS

```python
df["invoicedate"].max()
```

```python
today_date=dt.datetime(2011,12,11)
```

```python
type(today_date)
```

```python
rfm = df.groupby("customerid").agg({"invoicedate": lambda invoicedate: (today_date-invoicedate.max()).days, "invoice": lambda invoice: invoice.nunique(),"total_price": lambda total_price: total_price.sum()})
```

```python
rfm.head()
```

```python
rfm.sample(10)
```

```python
rfm.columns=["recency", "frequency","monetary"]
```

```python
rfm.describe().T
```

```python
rfm = rfm[rfm["monetary"]>0]
rfm.describe().T
```

```python
rfm["recency_score"] = pd.qcut(rfm["recency"], 5, labels=[5,4,3,2,1])
```

```python
rfm["frequency_score"]=pd.qcut(rfm["frequency"].rank(method="first"),5 , labels=[1,2,3,4,5])
```

```python
rfm["monetary_score"]=pd.qcut(rfm["monetary"],5, labels=[1,2,3,4,5])
```

```python
rfm["RFM_SCORE"]= (rfm["recency_score"].astype(str)+rfm["frequency_score"].astype(str))
```

```python
rfm.describe().T
```

```python
rfm[rfm["RFM_SCORE"]=="55"]
```

# Creating & Anlaysing RFM Segments

```python
seg_map = {
    r'[1-2][1-2]': 'hibernating',
    r'[1-2][3-4]': 'at_Risk',
    r'[1-2]5': 'cant_loose',
    r'3[1-2]': 'about_to_sleep',
    r'33': 'need_attention',
    r'[3-4][4-5]': 'loyal_customers',
    r'41': 'promising',
    r'51': 'new_customers',
    r'[4-5][2-3]': 'potential_loyalists',
    r'5[4-5]': 'champions'
}
```

```python
rfm["segment"]=rfm["RFM_SCORE"].replace(seg_map, regex=True)
```

```python
rfm[["segment", "recency", "frequency","monetary"]].groupby("segment").aggregate(["mean", "count"])
```
